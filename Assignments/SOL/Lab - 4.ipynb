```python
class Patient:
    def __init__(self, id, name, age, bloodgroup, next = None, prev = None):
        self.id, self.name, self.age, self.bloodgroup, self.next, self.prev = id, name, age, bloodgroup, next, prev

class WRM:
    def __init__(self):
        self.dh = Patient(None, None, None, None)
        self.dh.next = self.dh
        self.dh.prev = self.dh

0
    def RegisterPatient(self, id, name, age, bloodgroup):
        last_node = self.dh.prev
        p = Patient(id, name, age, bloodgroup, self.dh, last_node)
        last_node.next = p
        self.dh.prev = p

    def ServePatient(self):
        if self.dh.next==self.dh.prev==self.dh:
            print("No patient to be served\n")
        else:
            print(f"{self.dh.next.name} is served\n")
            self.dh.next = self.dh.next.next
            self.dh.next.prev = self.dh

    def CancelAll(self):
        if self.dh.next == self.dh.prev == self.dh:
            print("No appointments to cancel\n")
        else:
            self.dh.next = self.dh.prev = self.dh
            print("All appointments cancelled\n")

    def CanDoctorGoHome(self):
        if self.dh.next==self.dh.prev==self.dh:
            print("Yes\n")
        else:
            print("No\n")

    def ShowAllPatient(self):
        if self.dh.next==self.dh.prev==self.dh:
            print("No patient in the WRM\n")
        else:
            temp = self.dh.next
            while temp != self.dh:
                print(temp.name, end = " ")
                temp = temp.next
            print("\n")



dhdcl=WRM()
while True:
    print("==Choose an Option==\n1. RegisterPatient()\n2. ServePatient()\n3. CancelAll()\n4. CanDoctorGoHome()\n5. ShowAllPatient()\n6. Exit\n====================")

    inp = input("Enter your choice: ")

    if inp == "1":
        print("Executing RegisterPatient()...\n")
        id = input("Enter ID: ")
        name = input("Enter Name: ")
        age = input("Enter Age: ")
        bloodgroup = input("Enter Blood group: ")
        if id.isnumeric() and age.isnumeric and name.isalpha():
            dhdcl.RegisterPatient(id, name, age, bloodgroup)
            print("\nSuccess registering patient\n")
        else:
            print("\nFailure registering patient\n")

    elif inp == "2":
        print("Executing ServePatient()...\n")
        dhdcl.ServePatient()
    elif inp == "3":
        print("Executing CancelAll()...\n")
        dhdcl.CancelAll()
    elif inp == "4":
        print("Executing CanDoctorGoHome()...\n")
        dhdcl.CanDoctorGoHome()
    elif inp == "5":
        print("Executing ShowAllPatient()...\n")
        dhdcl.ShowAllPatient()
    elif inp == "6":
        print("\nThank You for Using Waiting Room Management System\nEXITING...")
        break
    else:
        print("\nNo Such Option\n")
```
